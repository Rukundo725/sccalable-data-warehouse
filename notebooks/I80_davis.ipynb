```python
from pandas import DataFrame
import pandas as pd
a_file = open("../data/I80_davis.txt") #openin text_file

# columns of our dataset
columns=['timestamp','ID','avg_speed','avg_flow','avg_occ','avg_freeflow_speed','samples_below_100pct_ff',
         'samples_below_95pct_ff','samples_below_90pct_ff','samples_below_85pct_ff','samples_below_80pct_ff',
         'samples_below_75pct_ff','samples_below_70pct_ff','samples_below_65pct_ff','samples_below_60pct_ff',
         'samples_below_55pct_ff','samples_below_50pct_ff','samples_below_45pct_ff','samples_below_40pct_ff',
         'samples_below_35pct_ff','samples_below_30pct_ff','samples_below_20pct_ff','samples_below_25pct_ff',
         'samples_below_15pct_ff','samples_below_10pct_ff' ]

# creating empty list to store corresponding values
timestamp=[]
ID=[]
avg_speed=[]
avg_flow=[]
avg_occ=[]
avg_freeflow_speed=[]
samples_below_100pct_ff=[]
samples_below_95pct_ff=[]
samples_below_90pct_ff=[]
samples_below_85pct_ff=[]
samples_below_80pct_ff=[]
samples_below_75pct_ff=[]
samples_below_70pct_ff=[]
samples_below_65pct_ff=[]
samples_below_60pct_ff=[]
samples_below_55pct_ff=[]
samples_below_50pct_ff=[]
samples_below_45pct_ff=[]
samples_below_40pct_ff=[]
samples_below_35pct_ff=[]
samples_below_30pct_ff=[]
samples_below_20pct_ff=[]
samples_below_25pct_ff=[]
samples_below_15pct_ff=[]
samples_below_10pct_ff=[]

# setting the limit of lines to excute because we have a huge that and it may cause our pc malfunctions
number_of_lines = 30000

for i in range(number_of_lines):
 

    line = a_file.readline() # reading each line
    a = line.split(',') # spliting elements of a line 
    
#     adding values to corresponding lists
    timestamp.append(a[0])
    ID.append(a[1])
    avg_speed.append(a[2])
    avg_flow.append(a[3])
    avg_occ.append(a[4])
    avg_freeflow_speed.append(a[5])
    samples_below_100pct_ff.append(a[6])
    samples_below_95pct_ff.append(a[7])
    samples_below_90pct_ff.append(a[8])
    samples_below_85pct_ff.append(a[9])
    samples_below_80pct_ff.append(a[10])
    samples_below_75pct_ff.append(a[11])
    samples_below_70pct_ff.append(a[12])
    samples_below_65pct_ff.append(a[13])
    samples_below_60pct_ff.append(a[14])
    samples_below_55pct_ff.append(a[15])
    samples_below_50pct_ff.append(a[16])
    samples_below_45pct_ff.append(a[17])
    samples_below_40pct_ff.append(a[18])
    samples_below_35pct_ff.append(a[19])
    samples_below_30pct_ff.append(a[20])
    samples_below_20pct_ff.append(a[21])
    samples_below_25pct_ff.append(a[22])
    samples_below_15pct_ff.append(a[23])
    samples_below_10pct_ff.append(a[24])

# zipping our list to data

data = zip(timestamp,ID,avg_speed,avg_flow,avg_occ,avg_freeflow_speed,samples_below_100pct_ff,samples_below_95pct_ff,
           samples_below_90pct_ff,samples_below_85pct_ff,samples_below_80pct_ff,samples_below_75pct_ff,samples_below_70pct_ff,
           samples_below_65pct_ff,samples_below_60pct_ff,samples_below_55pct_ff,samples_below_50pct_ff,samples_below_45pct_ff,
           samples_below_40pct_ff,samples_below_35pct_ff,samples_below_30pct_ff,samples_below_20pct_ff,samples_below_25pct_ff,
           samples_below_15pct_ff,samples_below_10pct_ff)

# creating a dataframe from our lists and columns

df = pd.DataFrame(data=data, columns=columns)

df.to_csv("../data/I80_davis.csv")
```
